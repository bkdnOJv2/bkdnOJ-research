```python
import json
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
CONTEST_ID = 20
```

```python
def html_standing(contest_id, page_index=1):
    URL = "https://codeforces.com/contest/{contestId}/standings/page/{pageIndex}"
    URL = URL.format(contestId=contest_id, pageIndex=page_index)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    return soup

def html_submission(contest_id, submission_id):
    URL = "https://codeforces.com/contest/{contestId}/submission/{submissionId}"
    URL = URL.format(contestId=contest_id, submissionId=submission_id)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    return soup
```

```python
npages = -1
page_index = 1
accepted_submission_id = []
```

```python
while True:
    soup = html_standing(CONTEST_ID, page_index)

    if npages == -1:
        npages = max(1, len(soup.select("[pageindex]")))
    
    for x in soup.select("[acceptedsubmissionid]"):
        accepted_submission_id.append(x['acceptedsubmissionid'])

    if npages == page_index:
        break
    page_index += 1
```

```python
def submission_obj(id):
    soup = html_submission(CONTEST_ID, id)
    info = soup.find_all("td")
    obj = {}
    obj['contest_id'] = CONTEST_ID
    obj['submission_id'] = info[0].text.strip()
    obj['author'] = info[1].text.strip()
    obj['problem'] = info[2].text.strip()
    obj['language'] = info[3].text.strip()
    obj['verdict'] = info[4].text.strip()
    obj['time'] = info[5].text.strip()
    obj['memory'] = info[6].text.strip()
    obj['sent'] = info[7].text.strip()
    obj['judged'] = info[8].text.strip()
    obj['source_code'] = soup.find(id='program-source-text').text

    return obj
```

```python
submissions = [submission_obj(id) for id in accepted_submission_id]
```

```python
df = pd.DataFrame(submissions)
```
